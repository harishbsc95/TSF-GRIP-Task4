# Import relevant libraries

```python
# Importing the libraries that require in this task
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Importing the dataset

```python
#To import and read data
data = pd.read_csv('globalterrorismdb_0718dist.csv')
data.head()
```

```python
#To check wether there are null values
data.isnull().sum()
```

```python
#To fill null values
missing_percentage = data.isnull().sum()*100/len(data)
missing_percentage
```

```python
#To plot a heatmap
plt.figure(figsize=(18,18))
sns.heatmap(data.isnull(), yticklabels=False, cbar=False, cmap="viridis")
```

```python
#To removie missing values
data.dropna(axis=1, inplace=True)
```

```python
#To display correlation
data.corr()
```

# Visualizing the data

```python
#To find we plot graph between no of attacks and attack year to find if there is increase in terrorism
plt.rcParams["figure.figsize"] = 15, 11
sns.countplot(x = data["iyear"])
plt.ylabel("Number of Attacks", fontsize=18)
plt.xticks(rotation=90)
plt.xlabel("Attack Year", fontsize=18)
plt.title("Increase in Terror", size=25, fontweight="bold")
```

```python
#To find sucess rate, we again
plt.rcParams["figure.figsize"] = 15, 11
sns.countplot(x = data["iyear"], hue="success", data=data)
plt.ylabel("Number of Attacks", fontsize=18)
plt.xticks(rotation=90)
plt.xlabel("Attack Year", fontsize=18)
plt.title("Increase in Terror", size=25, fontweight="bold")
```

```python
#Now we plot graph between number of attacks and attack region(comparing regionwise)
plt.rcParams["figure.figsize"] = 15, 11
sns.countplot(x = data["region_txt"])
plt.ylabel("Number of Attacks", fontsize=18)
plt.xticks(rotation=90)
plt.xlabel("Attack Year", fontsize=18)
plt.title("Attack Region", size=25, fontweight="bold")
```

```python
#To compare countrywise 
plt.rcParams["figure.figsize"] = 15, 11
sns.barplot(x = data["country_txt"].value_counts()[:20].index, y=data["country_txt"].value_counts()[:20].values)
plt.ylabel("Number of Attacks", fontsize=18)
plt.xticks(rotation=90)
plt.xlabel("Attack Region", fontsize=18)
plt.title("Country wise Attacks", size=25, fontweight="bold")
```

```python
#To compare attack types
plt.rcParams["figure.figsize"] = 15, 11
sns.barplot(x = data["attacktype1_txt"].value_counts()[:20].index, y=data["attacktype1_txt"].value_counts()[:20].values)
plt.ylabel("Total Attacks", fontsize=18)
plt.xticks(rotation=90)
plt.xlabel("Attack Type", fontsize=18)
plt.title("Attack Type", size=25, fontweight="bold")
```

```python
#To compare between groups
plt.rcParams["figure.figsize"] = 15, 11
sns.barplot(x = data["gname"].value_counts()[:20].index, y=data["gname"].value_counts()[:20].values)
plt.ylabel("Total Attack", fontsize=18)
plt.xticks(rotation=90)
plt.xlabel("Attack Group", fontsize=18)
plt.title("Attacks Group", size=25, fontweight="bold")
```
